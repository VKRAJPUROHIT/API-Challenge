```python
# Dependencies
import json
import matplotlib.pyplot as plt
from scipy import stats
import requests
import pandas as pd
from config import api_key
from citipy import citipy
import csv
import numpy as np
from random import uniform
import scipy.stats as st
from scipy.stats import linregress
from sklearn import datasets
```

```python
# Output File (CSV)
output_data_file = "output_data.csv"
```

```python
# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Build partial query URL
query_url = f"{url}appid={api_key}&units={units}&q="
```

```python
#make empty lists to populate the coordinates and cities  
coordinates=[]
cities = []
```

```python
# Create random coordinates
coordinates=[(uniform(-90.000,90.000),uniform(-180.000,180.000)) for i in range(0,2000)]
```

```python
# Identify nearest city for each lat, lng combination
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    
    # Replace spaces with %20 to create url correctly 
    city = city.replace(" ", "%20")      
 
    if city not in cities:
        cities.append(city)
print(len(cities))
```

```python
# Loop through the list of cities and perform a request for data on each
count=0
for city in cities:
#Try statement to populate when the values are found
# Sometimes values may not be returned     
    try:
        count=count+1
        response = requests.get(query_url + city).json()
        response1=requests.get(query_url+city)    
        cityid=response['id']
        print(f"weather check {count}:     CITY NAME:{city}      CITY ID:{cityid}      STATUS:{response1}")
    except KeyError:
        print(f"no city id")
print(f"End of weather check")        
        
```

```python
# set lists for the dataframe for weatherpy 
city_name=[]
lats = []
lngs = []
max_temps = []
humidity = []
cloudiness = []
wind_speeds = []
dates = []
countries=[]

# loops for creating dataframe columns
for city in cities:      
#Try statement to populate when the values are found
# Sometimes values may not be returned         
    try:
        response = requests.get(query_url + city).json()
        city_name.append(response['name'])
        lats.append(response['coord']['lat'])
        lngs.append(response['coord']['lon'])
        max_temps.append(response['main']['temp_max'])
        humidity.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        wind_speeds.append(response['wind']['speed'])         
        dates.append(response['dt'])    
        countries.append(response['sys']['country'])
    except KeyError:
        pass       
```

```python
#check the city number
print(len(city_name))
```

```python
# Assemble everything into a data frame
weather_df = pd.DataFrame({"City": city_name,
                           "City latitude": lats,
                           "City longitude": lngs,
                           "Max temperature": max_temps,                           
                           "Humidity": humidity,                           
                           "Cloud coverage": cloudiness,
                           "Wind speed": wind_speeds,                           
                           "City country": countries,
                           "City datetime": dates
                          })
weather_df.head()
```

```python
#recheck the number
print(len(weather_df))
```

```python
#Save data frame to a CSV
weather_df.to_csv(output_data_file)
```

## Linear Regression

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# Build a scatter plot for latitude vs temperature for northern hemisphere
plt.scatter(northern_df["City latitude"], northern_df["Max temperature"], marker="o")

# Incorporate the other graph properties
plt.title("Temperature in Northern Hemisphere(date of analysis:oct172020")
plt.ylabel("Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True)

# Add the linear regression equation and line to the scatter plot
x_values1 = northern_df["City latitude"]
y_values1 = northern_df["Max temperature"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values1, y_values1)
regress_values = np.array(x_values1) * slope + np.array(intercept)
line_eq1 =f'y={str(round(slope,2))}x + {str(round(intercept,2))}'
plt.plot(x_values1,regress_values,"r-")
plt.annotate(line_eq1,(6,10),fontsize=15,color="red")

#Calculate the correlation coefficient
correlation=st.pearsonr(x_values1,y_values1)

# Save the figure
plt.savefig("TemperatureInNorthernHemisphere.png")

print(f"The correlation coefficient between max temp and latitude in northern hemisphere is {round(correlation[0],2)}")
print(f"The equation of the regression line is: {line_eq1}")

# Show plot
plt.show()
```
